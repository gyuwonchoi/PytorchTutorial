### **TRANSFORM**
- feature 를 변경하기 위한 transform
- label을 변경하기 위한 targe_transform

```python
# PIL image type, Labels are integer

import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

ds=datasets.FashionMNIST(
    root='data',
    train=True,
    download= True,
    transform= ToTensor(), #img to tensor transform 
    target_transform=Lambda(lambda y: torch.zeros(10,dtype=torch.float).scatter_(0,torch.tensor(y), value=1)) #one-hot 
)
```

```python
src = torch.arange(1, 7).reshape((2, 3))
print(src)

index = torch.tensor([[0, 1, 2]])
result =torch.zeros(3, 3, dtype=src.dtype).scatter_(0, index, src)

print(result)
```
